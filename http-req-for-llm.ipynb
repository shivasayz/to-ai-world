```python
from openai import OpenAI

client = OpenAI(
    api_key = GROQ_API_KEY,
    base_url = "https://api.groq.com/openai/v1"   # root URL only
)

response = client.chat.completions.create(
    model="llama-3.1-8b-instant",   # corrected model name
    messages=[
        {
            "role": "system",
            "content": "You are a medical assistant. Answer only medical-related questions such as those about health, diseases, symptoms, treatments, medications, or healthcare. If a question is not related to medicine, politely respond that you cannot answer it because you only provide medical information."
        },
        {
            "role": "user",
            "content": "What is the Iran–Israel conflict?"
        }
    ]
)

print(response.choices[0].message.content)
```

```python
from openai import OpenAI

OLLAMA_BASE_URL = "http://localhost:11434/v1"

# Connect to local Ollama server
ollama = OpenAI(base_url=OLLAMA_BASE_URL, api_key="ollama")

response = ollama.chat.completions.create(
    model="gemma3:270m",
    messages=[{"role": "user", "content": "write something in telugu"}]
)

# Access the response correctly
res = response.choices[0].message.content
print(res)
```
